```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\kubatbekov.a\Downloads\исходные данные.xlsx', sheet_name='кейс 2')
```

```python
df = df.rename(columns={
    'Дата и время': 'datetime',
    'Сумма': 'amount',
    'Описание транзакции': 'description',
    'Получатель': 'recipient',
    'Статус платежа': 'status',
    'Место проведения': 'location'
})
```

```python
df.info()
```

```python
#Выше в инфо видим что колонка datetime строковый, переводим в тип данных datetime для дальнейшей работы
df['datetime'] = pd.to_datetime(df['datetime'])
df['month'] = df['datetime'].dt.to_period('M')
```
